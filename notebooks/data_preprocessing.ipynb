```python
import os
import cv2

# Define paths
VIDEO_DIR = "/Users/sid/Downloads/data"
OUTPUT_DIR = "/Users/sid/Downloads/data/frames"
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Process each category folder
for category in os.listdir(VIDEO_DIR):
    category_path = os.path.join(VIDEO_DIR, category)
    
    if os.path.isdir(category_path):  # Check if it's a folder
        output_category_path = os.path.join(OUTPUT_DIR, category)
        os.makedirs(output_category_path, exist_ok=True)

        for video_file in os.listdir(category_path):
            if video_file.endswith(".mp4"):  # Adjust for other formats if needed
                video_path = os.path.join(category_path, video_file)
                cap = cv2.VideoCapture(video_path)

                frame_count = 0
                frame_id = 0

                while cap.isOpened():
                    ret, frame = cap.read()
                    if not ret:
                        break  # Break if video ends

                    if frame_count % 30 == 0:  # 1 frame per second (assuming 30 FPS)
                        frame_filename = f"{video_file[:-4]}_frame{frame_id}.jpg"
                        frame_path = os.path.join(output_category_path, frame_filename)
                        cv2.imwrite(frame_path, frame)
                        frame_id += 1

                    frame_count += 1

                cap.release()

print("âœ… Frame extraction complete!")
```

```python
import cv2
import os

# --- PATHS ---
# Location of your dataset
DATA_DIR = "/Users/sid/Downloads/data"  
# Output folder for frames
OUTPUT_DIR = "/Users/sid/Downloads/data_frames"  

# Create the output directory if it doesn't exist
os.makedirs(OUTPUT_DIR, exist_ok=True)

# Get subdirectories in DATA_DIR (e.g. abuse, normal, robbery, shooting, etc.)
subdirs = [
    d for d in os.listdir(DATA_DIR) 
    if os.path.isdir(os.path.join(DATA_DIR, d))
]

# Define frame extraction rate (extract every 5th frame)
FRAME_STEP = 5  

# Loop over each subdirectory (activity class)
for activity in subdirs:
    activity_path = os.path.join(DATA_DIR, activity)
    activity_output_path = os.path.join(OUTPUT_DIR, activity)
    os.makedirs(activity_output_path, exist_ok=True)
    
    # Process each file in the activity folder
    for filename in os.listdir(activity_path):
        # Convert extension to lowercase and check if it's a video
        ext = os.path.splitext(filename)[1].lower()
        if ext not in [".mp4", ".avi", ".mov"]:
            # Skip csv or non-video files
            continue
        
        video_path = os.path.join(activity_path, filename)
        cap = cv2.VideoCapture(video_path)
        
        video_name = os.path.splitext(filename)[0]
        frame_count = 0
        saved_frame_count = 0

        while True:
            ret, frame = cap.read()
            if not ret:
                break  # No more frames left in the video
            
            if frame_count % FRAME_STEP == 0:
                frame_filename = f"{video_name}_frame{frame_count}.jpg"
                frame_filepath = os.path.join(activity_output_path, frame_filename)
                cv2.imwrite(frame_filepath, frame)
                saved_frame_count += 1
            
            frame_count += 1

        cap.release()
        print(f"[{activity}] Extracted {saved_frame_count} frames from: {filename}")

print("âœ… Extraction complete! Frames saved under:", OUTPUT_DIR)
```
